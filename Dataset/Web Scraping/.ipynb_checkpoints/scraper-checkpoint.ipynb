```python
import requests
from bs4 import BeautifulSoup

sitemap_url = "https://www.mentalhealth.org.uk/sitemap.xml"
r = requests.get(sitemap_url)
soup = BeautifulSoup(r.text, "xml")

urls = [loc.get_text() for loc in soup.find_all("loc")]
print("Total URLs:", len(urls))
print(urls[:10])
```

```python
#Only filter article

article_urls = [u for u in urls if "/explore-mental-health/" in u]
print(len(article_urls))
```

```python
#Create Scraping Functions to fetch html, extract titles and text, clean

import time, random
import pandas as pd
from bs4 import BeautifulSoup
import requests

headers = {
    "User-Agent": "FYPResearchBot/1.0 (+mailto:youremail@example.com)"
}

def scrape_article(url):
    """Scrape one article and return {title, body}."""
    r = requests.get(url, headers=headers, timeout=10)
    soup = BeautifulSoup(r.text, "lxml")
    
    # Extract the article title
    title_tag = soup.find("h1")
    title = title_tag.get_text(strip=True) if title_tag else "No title"
    
    # Extract main body text
    paragraphs = soup.find_all("p")
    body = " ".join(p.get_text(strip=True) for p in paragraphs)
    
    return {"url": url, "title": title, "body": body}
```

```python
#Loop through multiple URLs
articles = []

for i, url in enumerate(article_urls[:556]): #all articles
    try:
        data = scrape_article(url)
        articles.append(data)
        print(f"{i+1}/{len(article_urls)} âœ… {data['title'][:60]}")
        time.sleep(random.uniform(2, 5)) #polite delay
    except Exception as e:
        print(f"{i+1}/{len(article_urls)} âŒ Error on {url}: {e}")
    
    # Save partial progress
    if (i + 1) % 50 == 0:
        pd.DataFrame(articles).to_csv("partial_scrape.csv", index=False)
        print(f"ğŸ’¾ Saved {i+1} articles so far")

# Save final
pd.DataFrame(articles).to_csv("mental_health_articles.csv", index=False)
print("âœ… All 556 articles scraped and saved successfully!")
```
