```python
!pip install langdetect
#check what languages are inside
from langdetect import detect

import pandas as pd
df = pd.read_csv("mental_health_articles.csv")

langs = []
for text in df["body"].fillna(""):
    try:
        langs.append(detect(text))
    except:
        langs.append("unknown")

df["lang"] = langs
df["lang"].value_counts()
```

```python
#Keep malay and english only
df = df[df["lang"].isin(["en", "ms"])]
df.to_csv("mental_health_articles_filtered.csv", index=False)
print("✅ Filtered dataset saved!")
```

```python
#Check for duplicates

df.drop_duplicates(subset="url", inplace=True)
df.to_csv("mental_health_articles_final.csv", index=False)
```
